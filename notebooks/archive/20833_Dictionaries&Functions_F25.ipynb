## Dictionaries in Context

### Data Types
1. **int**: Integer (e.g., 5, -3, 0)
2. **float**: Floating-point number (e.g., 5.0, -3.14, 0.0)
3. **bool**: Boolean value (True or False)
4. **str**: String (e.g., "Hello", 'Python')


### Data Structures
1. **Lists**: One-dimensional, ordered collections where elements are indexed by their position (starting from 0).
2. **Dictionaries**: One-to-many mappings, where each unique key is associated with a single value. The values can be of any type, including other dictionaries, which can introduce more complex structures
3. **DataFrames**: Two-dimensional, structured data where you can index by both row and column.

## Iterate Through Dictionary

```python
for person in bellevue_people.keys():
    print(person)
```

```python
for person in bellevue_people.values():
  print(person['age'])
```

```python
for person in bellevue_people.values():
    if person['age'] > 20:
        name = person['name']
        age = person['age']
        print(f'{name} is more than 20 years old. They are {age}.')
    else:
      name = person['name']
      age = person['age']
      print(f"{name} doesn't meet criteria. They are {age}.")
```

## Define a Function

```python
# Functions are bundles of code that you plan to use a number of times
# Python has built-in functions we've already encountered: print() , len(), type()

# But you also have the ability to create (or "define") your own unique functions using the keyword "def"

def sing_personalized_happy_birthday(personalized_name):
    print("Happy Birthday to you")
    print("Happy Birthday to you")
    print(f"Happy Birthday dear {personalized_name}")
    print("Happy Birthday to you")


    return

# Here, "personalized_name" is the parameter requested by the function to complete its task
```

## Call a Function + Adding Parameters & Arguments

```python
# Using (or "calling") a function is simple. All you have to do is type the name of the function
# and provide specific content for the "parameter" (known as the "argument")

# In the function above "personalized_name" is the parameter.
# In the call of the function below, "Wesley" is the argument.

sing_personalized_happy_birthday("Wesley")
```

## Keyword Arguments

```python
def sing_personalized_happy_birthday(personalized_name, those_who_sing):
    print("Happy Birthday to you")
    print("Happy Birthday to you")
    print(f"Happy Birthday dear {personalized_name}")
    print("Happy Birthday to you")

    print(f"From {those_who_sing}")


    return
```

```python
sing_personalized_happy_birthday(input("Birthday kid? "), input("Your name? "))
```

## Return Values

```python
def add(a, b):
    result = a + b
    print(result)

sum = add(3, 5)
print(sum)
```

```python
def add(a, b):
    result = a + b
    return result

sum = add(5, 3)
print(sum)
```
