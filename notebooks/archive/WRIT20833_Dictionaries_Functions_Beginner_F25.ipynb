## Part 2: Your First Dictionary - A Complete Cultural Record

A **dictionary** is like a detailed information card about a cultural object. Instead of just storing a title, it stores multiple pieces of related information.

Think of it like this:
- A **list** is like a simple bookmark with just the title
- A **dictionary** is like a complete library catalog card with all the details

```python
# Our first dictionary - rich information about one book
evelyn_hugo = {
    "title": "The Seven Husbands of Evelyn Hugo",
    "author": "Taylor Jenkins Reid",
    "year": 2017,
    "genre": "Historical Fiction",
    "main_theme": "Celebrity, love, and identity",
    "why_significant": "Explores LGBTQ+ representation in historical Hollywood"
}

print("Complete book record:")
print(f"Title: {evelyn_hugo['title']}")
print(f"Author: {evelyn_hugo['author']}")
print(f"Published: {evelyn_hugo['year']}")
print(f"Genre: {evelyn_hugo['genre']}")
print(f"Theme: {evelyn_hugo['main_theme']}")
print(f"Significance: {evelyn_hugo['why_significant']}")
```

### Understanding Dictionary Structure

A dictionary has **keys** (the categories) and **values** (the information in those categories):

- `"title"` is a **key**, `"The Seven Husbands of Evelyn Hugo"` is its **value**
- `"author"` is a **key**, `"Taylor Jenkins Reid"` is its **value**
- And so on...

The **keys** tell us what type of information we're storing. The **values** are the actual information.

## Part 6: Introduction to Functions - Your First Analytical Tool

Imagine you want to get the basic information (title and author) from any book in your collection. Instead of writing the same code over and over, you can create a **function** - a reusable tool that does the work for you.

Think of a **function** like a recipe or a tool:
- You give it some ingredients (input)
- It follows steps to process those ingredients
- It gives you back the result (output)

```python
# Our first function - getting basic book information
def get_book_basics(book_dictionary):
    """
    This function takes a book dictionary and returns basic information.
    Input: A dictionary with book information
    Output: A nicely formatted string with title and author
    """
    title = book_dictionary["title"]
    author = book_dictionary["author"]
    
    basic_info = f"{title} by {author}"
    return basic_info

# Using our function
print("USING OUR FIRST FUNCTION")
print("=" * 25)

# Test with one book from our collection
americanah_basics = get_book_basics(identity_novels["americanah"])
print(f"Book 1: {americanah_basics}")

pachinko_basics = get_book_basics(identity_novels["pachinko"])
print(f"Book 2: {pachinko_basics}")

# We can use the same function for any book dictionary!
vuong_basics = get_book_basics(identity_novels["on_earth"])
print(f"Book 3: {vuong_basics}")
```

### Why Functions Are Powerful

Instead of writing the same code three times, we:
1. **Wrote the function once**
2. **Used it three times** with different books
3. **Got consistent results** every time

This is the beginning of **systematic analysis** - the foundation of digital humanities research!

```python
# Create a function that gets year and theme information
def get_book_context(book_dictionary):
    """
    This function takes a book dictionary and returns year and theme.
    Fill in the code below!
    """
    # Get the year from the book dictionary
    year = book_dictionary["year"]
    
    # Get the theme from the book dictionary
    theme = book_dictionary["theme"]
    
    # Create a formatted string with both pieces of information
    context_info = f"Published {year} - Theme: {theme}"
    
    # Return the result
    return context_info

# Test your function
print("TESTING YOUR FUNCTION")
print("=" * 22)

for book_key, book_info in identity_novels.items():
    title = book_info["title"]
    context = get_book_context(book_info)
    print(f"{title}: {context}")
```

## Part 7: Functions for Cultural Analysis

Let's create a more sophisticated function that can analyze patterns in our cultural collection. This is the kind of tool that digital humanities scholars use to study large collections of cultural objects.

```python
def analyze_publication_era(book_dictionary):
    """
    This function analyzes when a book was published and 
    categorizes it into a cultural era.
    """
    title = book_dictionary["title"]
    year = book_dictionary["year"]
    
    # Categorize by publication era
    if year < 2010:
        era = "Early 2000s"
        context = "Pre-social media literature"
    elif year < 2016:
        era = "Early 2010s"
        context = "Emerging diverse voices"
    else:
        era = "Late 2010s+"
        context = "Contemporary diverse literature boom"
    
    print(f"{title} ({year})")
    print(f"  Era: {era}")
    print(f"  Context: {context}")
    print()

# Use our analytical function
print("CULTURAL ERA ANALYSIS")
print("=" * 21)

for book_key, book_info in identity_novels.items():
    analyze_publication_era(book_info)
```

## Part 9: Finding Patterns Across Collections

One of the most powerful aspects of digital humanities is the ability to find patterns across large collections of cultural objects. Let's create a function that looks for patterns in our entire collection.

```python
def find_collection_patterns(cultural_collection, collection_name):
    """
    This function analyzes patterns across an entire collection.
    It shows how digital humanities scholars study cultural trends.
    """
    print(f"PATTERN ANALYSIS: {collection_name}")
    print("=" * 40)
    
    # Collect all the years
    years = []
    for item in cultural_collection.values():
        years.append(item["year"])
    
    # Find the range of years
    earliest_year = min(years)
    latest_year = max(years)
    
    print(f"Collection size: {len(cultural_collection)} items")
    print(f"Time span: {earliest_year} to {latest_year}")
    print(f"Covers {latest_year - earliest_year} years")
    
    # Look at genres
    genres = []
    for item in cultural_collection.values():
        genres.append(item["genre"])
    
    print(f"\nGenres represented:")
    for genre in set(genres):  # set() removes duplicates
        count = genres.count(genre)
        print(f"  - {genre}: {count} item(s)")
    
    print("\nCollection themes:")
    for item_key, item_info in cultural_collection.items():
        print(f"  - {item_info['title']}: {item_info['theme']}")

# Analyze our identity novels collection
find_collection_patterns(identity_novels, "Contemporary Identity Novels")

print("\n" + "=" * 50 + "\n")

# Analyze your collection
find_collection_patterns(my_collection, "My Personal Collection")
```
