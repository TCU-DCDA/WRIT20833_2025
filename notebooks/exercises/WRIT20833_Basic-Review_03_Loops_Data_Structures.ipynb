## Exercise 2: For Loops
Process lists with loops.

```python
# Loop through historical periods
periods = ["Ancient", "Medieval", "Renaissance", "Enlightenment", "Modern"]
years = ["3000 BCE - 500 CE", "500 - 1500 CE", "1300 - 1600", "1685 - 1815", "1800 - present"]

print("Historical Periods:")
for i in range(len(periods)):
    print(f"{i+1}. {periods[i]}: {years[i]}")

print("\nPeriod names only:")
for period in periods:
    print(f"- {period}")

# Count characters in each period name
print("\nName lengths:")
for period in periods:
    length = len(period)
    print(f"{period}: {length} characters")
```

## Exercise 3: List Processing
Filter and transform lists.

```python
# Process a list of literary genres
genres = ["poetry", "DRAMA", "fiction", "NON-FICTION", "biography", "ESSAY"]

print("Original genres:", genres)

# Standardize to title case
standardized = []
for genre in genres:
    standardized.append(genre.title())

print("Standardized:", standardized)

# Find long genre names (over 6 characters)
long_genres = []
for genre in standardized:
    if len(genre) > 6:
        long_genres.append(genre)

print("Long genre names:", long_genres)

# Count genres starting with vowels
vowel_count = 0
vowels = "AEIOU"
for genre in standardized:
    if genre[0] in vowels:
        vowel_count += 1

print(f"Genres starting with vowels: {vowel_count}")
```

## Exercise 4: Basic Dictionaries
Store related information together.

```python
# Create a book dictionary
book1 = {
    "title": "1984",
    "author": "George Orwell",
    "year": 1949,
    "pages": 328,
    "genre": "Dystopian Fiction"
}

print("Book Information:")
print(f"Title: {book1['title']}")
print(f"Author: {book1['author']}")
print(f"Published: {book1['year']}")
print(f"Pages: {book1['pages']}")
print(f"Genre: {book1['genre']}")

# Add new information
book1["rating"] = 4.5
book1["language"] = "English"

print(f"\nRating: {book1['rating']}/5")
print(f"Language: {book1['language']}")

# List all keys
print("\nAll information categories:")
for key in book1.keys():
    print(f"- {key}")
```

## Exercise 5: Dictionary Processing
Work with multiple dictionaries.

```python
# Create a collection of historical figures
figures = [
    {"name": "Leonardo da Vinci", "period": "Renaissance", "field": "Art/Science", "birth_year": 1452},
    {"name": "Shakespeare", "period": "Renaissance", "field": "Literature", "birth_year": 1564},
    {"name": "Marie Curie", "period": "Modern", "field": "Science", "birth_year": 1867},
    {"name": "Virginia Woolf", "period": "Modern", "field": "Literature", "birth_year": 1882}
]

print("Historical Figures Database:")
print("=" * 40)

for figure in figures:
    print(f"{figure['name']} ({figure['birth_year']})")
    print(f"Period: {figure['period']}, Field: {figure['field']}")
    print()

# Count by field
field_count = {}
for figure in figures:
    field = figure['field']
    if field in field_count:
        field_count[field] += 1
    else:
        field_count[field] = 1

print("Count by field:")
for field, count in field_count.items():
    print(f"{field}: {count}")
```

## Exercise 7: Text Analysis with Loops
Combine text processing with loops.

```python
# Analyze multiple texts
quotes = [
    "To be or not to be, that is the question",
    "I have a dream",
    "We hold these truths to be self-evident",
    "It was the best of times, it was the worst of times"
]

sources = ["Shakespeare", "Martin Luther King Jr.", "Declaration of Independence", "Charles Dickens"]

print("Quote Analysis:")
print("=" * 50)

for i in range(len(quotes)):
    quote = quotes[i]
    source = sources[i]
    
    word_count = len(quote.split())
    char_count = len(quote)
    has_comma = "," in quote
    
    print(f"Quote: \"{quote}\"")
    print(f"Source: {source}")
    print(f"Words: {word_count}, Characters: {char_count}")
    print(f"Contains comma: {has_comma}")
    
    if word_count > 10:
        print("Category: Long quote")
    else:
        print("Category: Short quote")
    
    print("-" * 30)
```
