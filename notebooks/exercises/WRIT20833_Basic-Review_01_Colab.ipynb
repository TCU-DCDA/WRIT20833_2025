## Exercise 1: Create Variables
Make variables for a book you are studying.

```python
# Create these variables about a book:
book_title = ""  # String: book title
author = ""  # String: author name
year_published = 0  # Integer: publication year
page_count = 0  # Integer: number of pages
is_fiction = True  # Boolean: True or False
rating = 0.0  # Float: rating out of 5.0

# Print all variables
print(f"Title: {book_title}")
print(f"Author: {author}")
print(f"Published: {year_published}")
print(f"Pages: {page_count}")
print(f"Fiction: {is_fiction}")
print(f"Rating: {rating}/5.0")
```

## Exercise 2: Check Data Types

```python
# Check the type of each variable
sample_text = "Hello World"
sample_number = 42
sample_decimal = 3.14
sample_boolean = True

print(type(sample_text))
print(type(sample_number))
print(type(sample_decimal))
print(type(sample_boolean))
```
