## Part 1: Understanding Found Data & Collection Ethics

Before diving into analysis, let's establish important context about **found data** and the ethical considerations that should guide your data collection and analysis.

### What is "Found Data"?
"Found data" refers to datasets that already exist in the world, waiting to be discovered and analyzed. For cultural research, this includes:
- **Public datasets** from Kaggle, government databases, academic repositories
- **Digital collections** from museums, libraries, archives
- **Scraped data** from websites (when ethically and legally permissible)
- **API data** from social media platforms, streaming services

### Data Collection Ethics & robots.txt
If your dataset was scraped from websites, it's crucial to understand **robots.txt** - the web's ethical guidelines for automated data collection.

**üìù Check if your data source respects robots.txt:**
- Add `/robots.txt` to any website URL (e.g., `https://example.com/robots.txt`)
- This file tells you what data collection is permitted
- **Always respect these guidelines** in your own research

**üí° Contemporary Issue: AI and Data Ethics**
The rise of AI training has created new debates about data collection ethics. Many websites now explicitly restrict their content from being used to train large language models. As cultural researchers, we must be even more thoughtful about:
- **Attribution**: Always credit your data sources
- **Scale**: Consider the impact of your data collection
- **Purpose**: Be transparent about your research goals
- **Consent**: Respect the intentions of data creators

```python
# Explore categorical/text columns
print("Text/Categorical columns analysis:")
for col in cultural_data.select_dtypes(include=['object']).columns:
    unique_count = cultural_data[col].nunique()
    print(f"\n{col}: {unique_count} unique values")
    if unique_count <= 10:
        print(f"  Values: {cultural_data[col].unique().tolist()}")
    else:
        print(f"  Sample values: {cultural_data[col].unique()[:5].tolist()}...")
```
